```python
from pymatgen.ext.matproj import MPRester
from pymatgen.phonon.plotter import PhononBSPlotter
import matplotlib.pyplot as plt
import numpy as np
```

```python
with MPRester("6FCK47cyF2mQPZ9CdW") as m:
    pbs = m.get_phonon_bandstructure_by_material_id("mp-581024")

pbsp = PhononBSPlotter(pbs)
info = pbsp.bs_plot_data()
dist = info['distances']
freq = info['frequency']
```

```python
pbsp.get_plot()

arr_1 = [freq[0][2][0], freq[0][2][1]]
dist_1 = [dist[0][0], dist[0][1]]

poly_1 = np.polyfit(dist_1, arr_1, 1) 
abs_1 = np.linspace(dist[0][0],dist[0][-1],300)
plt.plot(abs_1,np.polyval(poly_1, abs_1),'g')
    
arr_2 = [freq[3][1][-2], freq[3][1][-1]]
dist_2 = [dist[3][-2], dist[3][-1]]

poly_2 = np.polyfit(dist_2, arr_2, 1) 
abs_2 = np.linspace(dist[3][0],dist[3][-1],300)
plt.plot(abs_2,np.polyval(poly_2, abs_2),'r')

arr_3 = [freq[4][0][0], freq[4][0][1]]
dist_3 = [dist[4][0], dist[4][1]]

poly_3 = np.polyfit(dist_3, arr_3, 1) 
abs_3 = np.linspace(dist[4][0],dist[4][-1],300)
plt.plot(abs_3,np.polyval(poly_3, abs_3),'y')

plt.show()
```

```python
print("La vitesse du son dans la branche 1 dans la direction Γ-L vaut ", poly_3[0]*1e3, "m/s.")
print("La vitesse du son dans la branche 2 dans la direction Γ-K vaut ", -poly_2[0]*1e3, "m/s.")
print("La vitesse du son dans la branche 3 dans la direction Γ-X vaut ", poly_1[0]*1e3, "m/s.")
```
